SUMMARY OF EXPLORATORAY DATA ANALYSIS



The above dataset displays the data of 1000 students academic performanc(exam score) along with other factors which are:

- student_id	
- age	
- gender
- study_hours_per_day
- social_media_hours	
- netflix_hours	
- part_time_job	
- attendance_percentage	
- sleep_hours	
- diet_quality	
- exercise_frequency	
- parental_education_level	
- internet_quality	
- mental_health_rating	
- extracurricular_participation	

------------------------------------------------------------------------------------------------

- *The average class score is 69.6 out of 100*

- *The students age vary from 17 to 24 years. The data is almost evenly distributed among different ages.*

- *On an average the students are spending 3.5 hours on studying, 2.5 for social media, 1.8 for netflix  and 6.8 for sleeping.*

- *More than half of the class(53.7%) is neither involved in extracurricular activities nor doing any part time job suggesting ample amout of time for studies.However despite that the average exam score of this specific group remains closer to the class average*
  
- *In terms of physical health, majority of students follow fair to good amount of diet and exercise over 3 times a week. Students lying in both of these categories are 47.3% of the total .However the same cannot be commented on the mental health as the rating freuencies are almost evenly distributed*

- *Exam score follows a strong linear relationship with number of study hours. However there are certain outliers present which could be present due to influence of other factors*

- *People with average to poor internet quality are more likely to have higher mental health rating as well as higher exercise frequency than people with good internet quality*

- *Students whose parents education level is masters, are more likely to have a lower mental health rating.*
  
- *Average exam score of students with better mental health rating is more*
